```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class_Label','Alcohol','Malic Acid']
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python


# Assuming df is your DataFrame and contains columns 'Alcohol', 'Malic Acid', and 'Class_Label'
color_dict = {1: 'red', 2: 'blue', 3: 'green'}

# Correct usage of seaborn.scatterplot
sns.scatterplot(x='Alcohol', y='Malic Acid', hue='Class_Label', palette=color_dict, data=df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop('Class_Label',axis=1),df['Class_Label'],test_size=0.3,random_state=0)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler =  MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_train.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1) #Min-0   Max-1
```

```python
# Create a subplot with two columns
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Scatter plot for the original data
ax1.scatter(X_train['Alcohol'], X_train['Malic Acid'], c=y_train)
ax1.set_title("Before scaling")

# Scatter plot for the scaled data
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic Acid'], c=y_train)
ax2.set_title("After scaling")

# Show the plot
plt.show()
```

```python
fig , (ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))
#Before Scaling
ax1.set_title("Before scaling")
sns.kdeplot(X_train['Alcohol'],ax=ax1)
sns.kdeplot(X_train['Malic Acid'],ax=ax1)
#After Scaling
ax2.set_title("After scaling")
sns.kdeplot(X_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(X_train_scaled['Malic Acid'],ax=ax2)
plt.show() 
```
